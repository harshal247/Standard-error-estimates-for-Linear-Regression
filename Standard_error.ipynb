```python
import pandas as pd
import numpy as np
import math
from sklearn.utils import resample
from scipy.stats import norm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from scipy.stats import linregress
```

```python
df = pd.read_csv('amazon_bs_20102020.csv')
df = df.drop(columns=['Year', 'Rank','Book_Title','Author','Num_Customers_Rated'])
```

```python
df
```

```python
pop_estimate = np.mean(df['Price'])
```

```python
print("Estimate for the population mean: ",pop_estimate)
```

```python
std_dev = np.std(df['Price'])
```

```python
std_dev
```

```python
std_error = std_dev / math.sqrt(len(df))
```

```python
print("Standard Error: ",std_error)
```

```python
est_mean_list = []
price = df['Price']
for i in range(1000):
    df_resampled = resample(price, replace=True)
    est_mean_list.append(np.mean(df_resampled))
```

```python
est_mean_list
```

```python
std_of_est_mean = np.std(est_mean_list)
print("Standard error of 1000 sampled mean: ",std_of_est_mean)
```

```python
#95% confidence interval
ci1 = pop_estimate - 2 * std_error
ci2 = pop_estimate + 2 * std_error
print("95% Confidence Intervals : ",ci1,",",ci2)
```

```python
print("results through norm.interval method")
norm.interval(0.95, loc=pop_estimate, scale=std_error)
```

```python
rating = df['Rating']
```

```python
model = LinearRegression()
```

```python
rating = np.array(rating)
rating = rating.reshape(-1,1)
price = np.array(price)
price = price.reshape(-1,1)
```

```python
model_coef_list = []
model_intercept_list = []
for i in range(1000):
    rating_resampled = resample(rating, replace=True)
    price_resampled = resample(price, replace=True)
    
    model.fit(rating_resampled, price_resampled)
    model_coef_list.append(model.coef_)
    model_intercept_list.append(model.intercept_)
```

```python
print("Coefficient Standard Error:",np.std(model_coef_list))
```

```python
print("Intercept Standard Error",np.std(model_intercept_list))
```
